#Seaborn

1.distplot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
database=sb.load_dataset('diamonds')
print(database)
```

```python
sb.displot(database['table'])
plt.show()
```

2.jointplot

```python
sb.jointplot(x='cut',y='depth',data=database)
plt.show()
```

3.catplot

```python
sb.catplot(x='cut',y='clarity',data=database,kind='strip')
plt.show()
```

4.FacetGrid and map

```python
graph=sb.FacetGrid(database,col='cut')
graph.map(plt.scatter,'table','price')
plt.show()
```
