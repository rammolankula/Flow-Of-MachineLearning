#BMH

```python
import matplotlib.pyplot as plt
from matplotlib import style
style.use('bmh')
x=[1.2,2.3]
y=[2.3,4.5]
x2=[2,4]
y2=[6,4]
plt.title('test')
plt.xlabel('x label value')
plt.ylabel('y label value')
plt.plot(x,y)
plt.plot(x2,y2)
plt.show()
```

#bar graph and classic style

```python
import matplotlib.pyplot as plt
from matplotlib import style
style.use('classic')
x1=[2,4,6]
y1=[3,6,9]


plt.bar(x1,y1)

plt.title('test')
plt.xlabel('x test name')
plt.ylabel('y test name')
plt.show()
```

#bmh style and using histogramgrapgh numbers to jump

```python
import matplotlib.pyplot as plt
from matplotlib import style

style.use('bmh')

numbers=[0,12,11,12,13,16,18,19,20,22,26,28,29,34,37,39,42,43,46,57,68,14]
Jumps=[0,2,5,10,25,45,59,80]

plt.title('test of this')
plt.xlabel('the value of x')
plt.ylabel(' the value of y')
plt.hist(numbers,Jumps,histtype='bar')
plt.show()
```

#pie chart an bmh style

```python
import matplotlib.pyplot as plt
from matplotlib import style
style.use('bmh')
food=['pizza','ice cream','fulka']
sales=[30,10,20]
color=['red','white','brown']
plt.title('Example of huge sale food')
plt.pie(sales,labels=food,colors=color)
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import style
style.use('classic')

food=['pizza','ice cream','Rice']
sale=[20,20,20]
color=['red','white','brown']

plt.pie(sale,labels=food,colors=color)
plt.title('Example of common sale food')
plt.show()
```
